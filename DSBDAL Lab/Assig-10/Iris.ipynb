```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IRIS.csv')
```

```python
df.head()
```

```python
df.dtypes
```

Histogram for each feature in the dataset

```python
df.hist(edgecolor='black', linewidth=1)
fig=plt.gcf()
fig.set_size_inches(10, 8)
plt.show()
```

```python
sns.histplot(df, x='sepal_length', y='petal_length', hue='species')
```

```python
sns.histplot(df, x='sepal_width', y='petal_width', hue='species')
```

```python
sns.histplot(df, x='sepal_length', y='species')
```

```python
sns.histplot(df, x='sepal_length', y='petal_width')
```

```python
sns.histplot(df, x='sepal_width', y='species', color="red")
```

Box plot for each feature in the dataset

```python
df.boxplot(color='black')
fig=plt.gcf()
fig.set_size_inches(8, 6)
plt.show()
```

```python
sns.boxplot(x=df['petal_length'])
```

```python
sns.boxplot(x=df['sepal_length'])
```
