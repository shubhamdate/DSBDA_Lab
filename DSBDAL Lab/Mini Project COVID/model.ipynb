# COVID Vaccination Analysis

## Import Necessary Liberaires

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 25)
```

## Import Dataset

```python
dataset = pd.read_csv('covid_vaccine_statewise.csv')
dataset.head()
```

## Data Analysis

```python
# Shape of Dataset
print("Number of Rows:-> ", dataset.shape[0])
print("Number of Columns:-> ", dataset.shape[1])
```

```python
dataset.info()
```

```python
# a. Describe Dataset
dataset.describe()
```

```python
dataset = dataset[dataset['State'] != 'India']
```

## Check for NaN values

```python
dataset.isnull().sum()
```

```python
total_values = dataset.shape[0]
first_dose_nan = dataset['First Dose Administered'].isnull().sum()
print("Percentage of NaN in First Dose Vaccination:-> ", round(100-(total_values-first_dose_nan)/total_values*100, 2), "%")
```

```python
total_values = dataset.shape[0]
second_dose_nan = dataset['Second Dose Administered'].isnull().sum()
print("Percentage of NaN in Second Dose Vaccination:-> ", round(100-(total_values-second_dose_nan)/total_values*100, 2), "%")
```

```python
fd_mean = dataset['First Dose Administered'].mean()
sd_mean = dataset['Second Dose Administered'].mean()

dataset['First Dose Administered'].fillna(fd_mean, inplace=True)
dataset['Second Dose Administered'].fillna(sd_mean, inplace=True)
```

```python
dataset.isnull().sum()
```

```python
stateDataset = dataset[['State', 'First Dose Administered', 'Second Dose Administered']].groupby('State').sum()
```

```python
# b. Number of persons statewise vaccinated for first dose in India
# c. Number of persons statewise vaccinated for second dose in India

stateDataset
```

## Data Visualization

```python
stateDataset['Total Dose Administered'] = stateDataset['First Dose Administered']+stateDataset['Second Dose Administered']
```

```python
sns.histplot(data=stateDataset, x='State', y='Total Dose Administered')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(data=stateDataset, x='Total Dose Administered')
plt.show()
```

```python
date_dataset = dataset[['Updated On', 'First Dose Administered', 'Second Dose Administered']].groupby('Updated On').sum()
```

```python
male_nan = dataset['Male(Individuals Vaccinated)'].isnull().sum()
print("Percentage of NaN values in male vaccinated feature: ", 100-((total_values-male_nan)/total_values)*100, "%")
```

```python
# Filling NaN values
male_mean = dataset['Male (Doses Administered)'].mean()
female_mean = dataset['Female (Doses Administered)'].mean()

dataset['Male (Doses Administered)'].fillna(male_mean, inplace=True)
dataset['Female (Doses Administered)'].fillna(female_mean, inplace=True)
```

```python
# d. Number of Male Dose Administered
# e. Number of Female Dose Administered
print("Male Dose Administered : ", dataset['Male (Doses Administered)'].sum())
print("Female Dose Administered : ", dataset['Female (Doses Administered)'].sum())
```
