```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.shape
```

```python
print("Number of peoples survived:-> ", dataset['survived'].value_counts()[1])

print("Number of peoples Not survived:-> ", dataset['survived'].value_counts()[0])
```

```python
sns.countplot(data = dataset, x='survived')
plt.show()
```

```python
sns.histplot(dataset['fare'])
plt.show()
```

```python
sns.boxplot(data=dataset, x='age', y='sex', hue='survived')
plt.show()
```

```python
sns.pairplot(dataset)
plt.show()
```

```python
gender = dataset['sex'].value_counts()
```

```python
data = [gender['male'], gender['female']]
labels = ['Male', 'Female']

plt.pie(data, labels = labels, autopct='%.0f%%')
plt.show()
```
